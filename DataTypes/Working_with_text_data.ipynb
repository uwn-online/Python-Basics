```python
products = ['Milk', 'Sugar', 'Biscuits', 'Eggs']
exp_sales = [10000, 230000, 40000, 50000, 120000]
```

```python
time_horizon = (1, 3, 12)
```

```python
for i in products:
    for j in exp_sales:
        for k in time_horizon:
            print([i, j*k])
```
